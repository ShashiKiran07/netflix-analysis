```python
# def country_count(state):
#     count = 0
#     for i in netflix['country']:
#         if state in i:
#             count += 1
#     return count

# brazil = country_count('Brazil')
# print(brazil)
def num_times(state):
    count = 0
    for i in netflix['country']:
        l = i.split(',')
        if state in l:
            count += 1
        else:
            pass
    return count
greece = num_times('Greece')
india = num_times('India')
turkey = num_times('Turkey')
united_states = num_times('United States')
united_kingdom = num_times('United Kingdom')
canada = num_times('Canada')
mexico = num_times('Mexico')
korea = num_times('South Korea')
    
```

```python
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(5)
ax = fig.add_axes([0,0,1,1])
countries = [greece, india, turkey, united_states, united_kingdom, canada, mexico, korea]
names = ['Greece', 'India', 'Turkey', 'United States', 'United Kingdom', 'Canada', 'Mexico', 'Korea']
ax.bar(names,countries)
plt.show()
```

```python
movies_by_years = netflix.loc[netflix.country == 'United States'].groupby('release_year').title.count()
print(movies_by_years)
```

```python
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)
movies_by_years.plot.bar()
plt.show()
```

```python
movies_or_series = netflix.loc[netflix.country == 'United States'].groupby(['release_year','type']).title.count()
print(movies_or_series)
```

```python
fig = plt.figure()
fig.set_figwidth(20)
fig.set_figheight(10)
movies_or_series.plot.bar()
plt.show()
```
